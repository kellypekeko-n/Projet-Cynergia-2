```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

print("1. Chargement du dataset...")
df = pd.read_csv("Dataset.csv")

print("2. Extraction des caractéristiques (Features)...")
# On isole les colonnes numériques de 'duration' à 'rAckDelayAvg'
X = df.loc[:, 'duration':'rAckDelayAvg']

# Nettoyage de sécurité : remplace les valeurs manquantes (NaN) ou infinies s'il y en a
X = X.replace([np.inf, -np.inf], np.nan).fillna(0)

print("3. Encodage des étiquettes (Labels)...")
y_raw = df['NST_M_Label']
encoder = LabelEncoder()
y = encoder.fit_transform(y_raw)

print("4. Normalisation des données...")
# Étape essentielle pour que le Réseau de Neurones fonctionne correctement
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("5. Division des données (80% Entraînement / 20% Test)...")
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("\n=== [SUCCÈS] TOUTES LES DONNÉES SONT PRÊTES ! ===")
print(f"Lignes d'entraînement : {X_train.shape[0]} | Lignes de test : {X_test.shape[0]}")
print("Classes cibles identifiées :", dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

# --- 1. INITIALISATION DES ALGORITHMES ---
print("Initialisation des 3 algorithmes d'IA...")
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=100, random_state=42, eval_metric='logloss'),
    "Réseau de Neurones (MLP)": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=200, random_state=42)
}

# --- 2. ENTRAÎNEMENT ET PRÉDICTION ---
results = {}

for name, model in models.items():
    print(f"\n--- Entraînement du modèle : {name} ---")
    model.fit(X_train, y_train)
    
    print(f"Évaluation de : {name}...")
    y_pred = model.predict(X_test)
    
    # Stockage des prédictions et scores
    results[name] = {
        "accuracy": accuracy_score(y_test, y_pred),
        "report": classification_report(y_test, y_pred, target_names=encoder.classes_, output_dict=True)
    }
    print(f"-> Terminé ! Précision globale : {results[name]['accuracy']*100:.2f}%")

print("\n=== [SUCCÈS] L'ENTRAÎNEMENT COMPARATIF EST TERMINÉ ! ===")
```
